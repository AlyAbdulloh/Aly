```python
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow # for image display
from skimage import io
from skimage import transform
from PIL import Image
import matplotlib.pylab as plt
```

```python
img_1 = cv.imread('/content/sample_data/KTP.png')
img_2 = cv.cvtColor(img_1, cv.COLOR_BGR2RGB)
# img_2 = cv.imread('/content/sample_data/KTP.png')
cv2_imshow(img_1)
plt.imshow(img_2)
```

TUTUP BAGIAN

```python
img_1s = np.copy(img_1)
#tutup bagian NIK
for y in range (110,280):
  for x in range (50,70):
    img_1s[x,y] = [0,255,255]

img_1ss = np.copy(img_1s)
#tutup bagian nama
for y in range (122,155):
  for x in range (75,85):
    img_1ss[x,y] = [0,0,0]

img_1sss = np.copy(img_1ss)
#tutup foto ktp
for y in range (300,390):
  for x in range(60,170):
    img_1sss[x,y] = [255,255,255]


final_frame = cv.hconcat((img_1, img_1sss))
cv2_imshow(final_frame)
```
